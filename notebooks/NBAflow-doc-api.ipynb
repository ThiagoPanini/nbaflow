## Dados de Jogadores

```python
# Adicionando diret√≥rio do ambiente virtual como uma vari√°vel de ambiente para importa√ß√£o dos pacotes
import sys
sys.path.append('/home/paninit/python-venvs/nbaflow-venv/lib/python3.8/site-packages')

# Importando a biblioteca
from nba_api.stats.static import players
import pandas as pd
pd.set_option('display.max_columns', 500)

# Retornando todos os jogadores dispon√≠veis
df_players = pd.DataFrame(players.get_players())
print(f'Dimens√µes do DataFrame retornado: {df_players.shape}')
df_players.head()
```

Analisando o bloco de c√≥digo acima, nota-se que a biblioteca `nba-api`, atrav√©s de seu m√≥dulo `stats.static`, fornece o objeto `players` que, por sua vez, √© respons√°vel por entregar alguns m√©todos importantes relacionados a an√°lise pontual de jogadores da NBA. Exemplo disso √© o m√©todo `get_players()` com o objetivo de retornar todos os jogadores presentes nas bases de dados da NBA em um formato de dicion√°rio. Atrav√©s do `pandas`, pode-se transformar esse resultado em um objeto DataFrame para uma maior facilidade de visualiza√ß√£o e poss√≠vel tratamento.

Ainda de acordo com a documenta√ß√£o, √© poss√≠vel ainda utilizar diversos m√©todos para o retorno de informa√ß√µes de jogadores utilizando o objeto `players`. Um deles, aparentemente relevante, permite retornar apenas os jogadores ativos (apesar dessa possibilidade existir tamb√©m com o `get_players()` e a aplica√ß√£o de um filtro na coluna *is_active*). Vejamos o resultado de uma requisi√ß√£o de somente jogadores ativos atrav√©s do m√©todo `get_active_players()` do mesmo objeto.

```python
# Retornando apenas jogadores ativos
df_active_players = pd.DataFrame(players.get_active_players())
print(f'Dimens√µes da base de jogadores ativos pelo m√©todo "get_active_players()": {df_active_players.shape}')
print(f'Dimens√µes da base total de jogadores filtrando apenas jogadores ativos: {df_players.query("is_active==True").shape}')
df_active_players.head()
```

Perfeito! Aparentemente os m√©todos funcionam perfeitamente bem dentro do prop√≥sito de extrair jogadores ativos na liga. Ao todo, s√£o 519 registros de jogadores. Como um pr√≥ximo desafio, pretende-se utilizar informa√ß√µes como o _id_ de um determinado jogador para extrair informa√ß√µes mais detalhadas sobre o mesmo.

Para isso, ser√° utilizado o objeto `commonplayerinfo` do m√≥dulo `stats.endpoints`.

```python
# Importando objeto
from nba_api.stats.endpoints import commonplayerinfo

# Retornando um id espec√≠fico de jogador
dame_id = df_active_players.query('full_name == "Damian Lillard"')['id'].values[0]

# Criando objeto da reqisi√ß√£o
player_info = commonplayerinfo.CommonPlayerInfo(player_id=dame_id)
player_info
```

A partir deste ponto, o objeto `player_info` pode ser acessado para retornar diferentes informa√ß√µes em diferentes requisi√ß√µes. Neste caso, existem datasets que podem ser acessados a partir da requisi√ß√£o de diferentes atributos:

* `player_headline_stats`: para retornar um cabe√ßalho estat√≠stico do jogador
* `common_player_info`: para retornar dados completos relacionados ao jogador
* `available_seasons`: para retornar informa√ß√µes de temporadas v√°lidas do jogador em quest√£o

Adicionalmente, ap√≥s chamar um dos m√©todos acima a partir do objeto `player_info`, √© poss√≠vel aplicar outros m√©todos respons√°veis por direcionar o formato do objeto de sa√≠da dos dados a ser analisado. Assim, √© poss√≠vel utilizar os m√©todos:

* `get_response()`: para retornar a resposta bruta da requisi√ß√£o
* `get_json()`: para retornar os dados em um formato string simulando um JSON
* `get_dict()`: para retornar os dados em um formato de dicion√°rio python
* `get_data_frame()`: para retornar os dados em um formato DataFrame do pandas

Consolidando exemplos pr√°ticos de tudo o que foi dito acima, os c√≥digos abaixo navegam por algumas das combina√ß√µes dispon√≠veis para retornar informa√ß√µes resumidas, completas e de temporadas relacionadas ao jogador em quest√£o (par√¢metro id passado como entrada).

```python
# Retornando resumo estat√≠stico em formato de json de Damian Lillard
dame_headline_stats = player_info.player_headline_stats.get_json()
dame_headline_stats
```

```python
# Retornando informa√ß√µes completas de Damian Lillard em um formato DataFrame
dame_info = player_info.common_player_info.get_data_frame()
dame_info
```

```python
# Retornando temporadas dispon√≠veis de Damian Lillard
dame_seasons = player_info.available_seasons.get_dict()['data'][:5]
print('Obs: mostrando apenas os 5 primeiros elementos')
dame_seasons
```

## Dados de Arremessos

Explorando a documenta√ß√£o fornecida em mais detalhes, √© poss√≠vel encontrar uma s√©rie de [endpoints](https://github.com/swar/nba_api/tree/master/docs/nba_api/stats/endpoints) com diferentes contextos de dados. Buscando informa√ß√µes relacionadas a arremessos de jogadores, foi poss√≠vel chegar at√© o endpoint [`shotchartdetail`](https://github.com/swar/nba_api/blob/master/docs/nba_api/stats/endpoints/shotchartdetail.md) com uma capacidade altamente promissora para o retorno de dados estat√≠sticos de jogadores espec√≠ficos, partida a partida, de acordo com par√¢metros de entrada passados como argumento.

```python
# Importando m√≥dulo
from nba_api.stats.endpoints import shotchartdetail

# Retornando dados de arremessos de Damian Lillard no Portland TrailBlazers
blazers_id = dame_info['TEAM_ID'].values[0]
dame_blazers_datasets = shotchartdetail.ShotChartDetail(player_id=dame_id, team_id=blazers_id)
print(f'Tipo do objeto retornado: {type(dame_blazers_datasets)}')
```

Ao instanciar a classe `ShotChartDetail` passando o id de um jogador e o id de um time, √© poss√≠vel retornar um objeto contendo informa√ß√µes √∫teis a respeito dessa combina√ß√£o jogador/time. Assim como nos exemplos mencionados na sess√£o anterior, o resultado permite executar alguns m√©todos espec√≠ficos para o retorno dos dados em si.

Nesse caso, √© poss√≠vel executar o m√©todo `get_data_frames()` para retornar todas as bases de dados presentes no objeto resultante da requisi√ß√£o. Na pr√°tica, investigando um pouco em detalhes, √© poss√≠vel pontuar que a resposta proporcionada pela consulta em `ShotChartDetail` possui duas principais bases, sendo estas acess√≠veis pelos seguintes atributos:

* `short_chart_detail`: retorna informa√ß√µes granulares de todos os arremessos realizados pelo jogador especificado no time especificado em todas as partidas dispon√≠veis
* `league_averages`: retorna m√©dias gerais, por tipo de arremesso, da liga como um todo

Lembrando que, assim como nos casos anteriores, √© preciso executar m√©todos referentes a especifica√ß√£o do retorno dos dados relacionados (DataFrame, json ou dicion√°rio).

```python
# Retornando dados de todos os arremessos de Damian Lillard pelo Portland
dame_blazers_shots = dame_blazers_datasets.shot_chart_detail.get_data_frame()
print(f'Dimens√µes da base de arremessos de Damian no Portland: {dame_blazers_shots.shape}')
dame_blazers_shots.head()
```

```python
# Retornando m√©dias gerais da liga
league_averages = dame_blazers_datasets.league_averages.get_data_frame()
league_averages.head()
```

Ainda que seja poss√≠vel retornar dados relevantes de arremessos de um jogador espec√≠fico utilizando a classe `ShortChartDetail`, √© necess√°rio fornecer uma combina√ß√£o _jogador_ e _time_, fazendo assim com que casos onde jogadores que atuaram em mais de um time na carreira pudesse ser enviesado e dif√≠cil de gerenciar nesse formato de requisi√ß√£o.

Para contornar essa exig√™ncia, √© poss√≠vel passar, no par√¢metro da requisi√ß√£o, ou seja, como argumentos de inicializa√ß√£o da classe, o valor `team_id=0`, permitindo assim com que o retorno dado contenha todos os dados de um determinado jogador para todos os times que este tenha passado na carreira (desde que outros par√¢metros de filtro, como season, por exemplo, n√£o estejam definidos previamente). No c√≥digo abaixo, ser√° proposto o retorno de todos os dados de LeBron James.

```python
# Retornando dados de LeBron James
lebron_id = players.find_players_by_first_name('Lebron')[0]['id']
lebron_data = shotchartdetail.ShotChartDetail(player_id=lebron_id, team_id=0)
lebron_shots = lebron_data.shot_chart_detail.get_data_frame()

print(f'Dimens√µes da base de arremessos de LeBron James na liga: {lebron_shots.shape}')
lebron_shots.head()
```

```python
# Adicionando diret√≥rio do ambiente virtual como uma vari√°vel de ambiente para importa√ß√£o dos pacotes
import sys
sys.path.append('/home/paninit/python-venvs/nbaflow-venv/lib/python3.8/site-packages')

# Importando recursos
from nba_api.stats.static import players
from nba_api.stats.endpoints import commonplayerinfo, playergamelog
import os
import pandas as pd
import logging
from datetime import datetime

# Configura√ß√£o adicional do pandas
pd.set_option('display.max_columns', 500)

# Definindo fun√ß√£o para configurar objeto de log do c√≥digo
def log_config(logger, level=logging.DEBUG, 
               log_format='%(levelname)s;%(asctime)s;%(filename)s;%(module)s;%(lineno)d;%(message)s',
               log_filepath=os.path.join(os.getcwd(), 'exec_log/execution_log.log'),
               flag_file_handler=False, flag_stream_handler=True, filemode='a'):
    """
    Fun√ß√£o que recebe um objeto logging e aplica configura√ß√µes b√°sicas ao mesmo
    
    Par√¢metros
    ----------
    :param logger: objeto logger criado no escopo do m√≥dulo [type: logging.getLogger()]
    :param level: level do objeto logger criado [type: level, default=logging.DEBUG]
    :param log_format: formato do log a ser armazenado [type: string]
    :param log_filepath: caminho onde o arquivo .log ser√° armazenado 
        [type: string, default='exec_log/execution_log.log']
    :param flag_file_handler: define se ser√° criado um arquivo de armazenamento de log
        [type: bool, default=False]
    :param flag_stream_handler: define se as mensagens de log ser√£o mostradas na tela
        [type: bool, default=True]
    :param filemode: tipo de escrita no arquivo de log [type: string, default='a' (append)]
    
    Retorno
    -------
    :return logger: objeto logger pr√©-configurado
    """

    # Setting level for the logger object
    logger.setLevel(level)

    # Creating a formatter
    formatter = logging.Formatter(log_format, datefmt='%Y-%m-%d %H:%M:%S')

    # Creating handlers
    if flag_file_handler:
        log_path = '/'.join(log_filepath.split('/')[:-1])
        if not isdir(log_path):
            makedirs(log_path)

        # Adding file_handler
        file_handler = logging.FileHandler(log_filepath, mode=filemode, encoding='utf-8')
        file_handler.setFormatter(formatter)
        logger.addHandler(file_handler)

    if flag_stream_handler:
        # Adding stream_handler
        stream_handler = logging.StreamHandler()
        stream_handler.setFormatter(formatter)    
        logger.addHandler(stream_handler)

    return logger

# Instanciando e configurando objeto de log
logger = logging.getLogger('nbaflow-notebook')
logger = log_config(logger)
```

```python
# Classe para gerenciamento e extra√ß√£o de hist√≥rico de partidas de jogadores da NBA
class NBAGamelog:
    """
    Classe respons√°vel por representar um objeto de extra√ß√£o de
    hist√≥rico de partidas de jogadores da NBA. Em seu conjunto de
    m√©todos e atributos, √© poss√≠vel gerenciar requisi√ß√µes √† nba-api
    dentro de um contexto de dados de partidas de jogadores espec√≠ficos
    e em temporadas espec√≠ficas (Regular ou Playoffs).
    
    A constru√ß√£o dos m√©todos dessa classe foi pensada de tal modo a 
    proporcionar que o usu√°rio final realize requisi√ß√µes sob demanda
    de acordo com o prop√≥sito definido, seja este pautado pela an√°lise
    pontual de partidas de um √∫nico jogador em uma √∫nica temporada, ou 
    ent√£o por um processo mais robusto para extrair todo o hist√≥rico
    de partidas dispon√≠veis de todos os jogadores ativos da liga.
    
    Al√©m disso, entre as funcionalidades implementadas em m√©todos
    mais robustos que envolvem um grande n√∫mero de requisi√ß√µes, foram
    consideradas l√≥gicas que permitem o reprocessamento din√¢mico de 
    requisi√ß√µes com status de falha a partir de um incremento linear
    no timeout originalmente definido. Em outras palavras, ao executar
    m√©todos respons√°veis pela extra√ß√£o de um grande n√∫mero de partidas,
    em caso de eventuais falhas nas requisi√ß√µes, os dados s√£o armazenados
    temporariamente em um atributo da classe para posterior reprocessamento,
    caso esta seja a configura√ß√£o definida pelo usu√°rio. Dessa forma, √© 
    poss√≠vel parametrizar os m√©todos para que seja garantida uma efici√™ncia
    de 100% nas requisi√ß√µes realizadas √† api da NBA.
    
    Atributos da classe
    -------------------
    :attr verbose: 
        Define a verbosidade dos logs ao longo dos m√©todos. Quanto maior seu
        valor, mais logs ser√£o mostrados ao usu√°rio e eventualmente armazenados.
        [type: int, default=1]
        
    :attr error_cols:
        Vari√°veis que guiam e definem as requisi√ß√µes realizadas com status
        de falha. Basicamente, definem quais informa√ß√µes dessas requisi√ß√µes
        ser√£o armazenadas de modo a propor um reprocessamento futuro.
        [type: list, default=['player_id', 'player_name', 'season', 'season_type', 'timeout']]
        
    :attr error_data:
        Consiste em um objeto DataFrame respons√°vel por alocar requisi√ß√µes
        realizadas com status de falha. Suas colunas s√£o definidas pelo 
        atributo self.error_cols.
        [type: pd.DataFrame]
        
    :attr all_players:
        Base contendo informa√ß√µes b√°sicas de todos os jogadores da NBA
        extra√≠da diretamente do m√©todo est√°tico players.get_players()
        importado do m√≥dulo nba_api.stats.static
        [type: pd.DataFrame]
        
    :attr active_pƒ∫ayers:
        Base contendo informa√ß√µes b√°sicas de jogadores ativos na NBA
        extra√≠da diretamente do m√©todo est√°tico players.get_active_players()
        importado do m√≥dulo nba_api.stats.static
        [type: pd.DataFrame]
    """
    
    def __init__(self, verbose=1):
        self.error_cols = ['player_id', 'player_name', 'season', 'season_type', 'timeout']
        self.error_data = pd.DataFrame(columns=self.error_cols)
        self.all_players = pd.DataFrame(players.get_players())
        self.active_players = pd.DataFrame(players.get_active_players())
        self.verbose = verbose
         
    # Auxilia na estrutura√ß√£o de requisi√ß√µes com falhas para futuro armazenamento
    def structure_error_data(self, player_id, player_name, season, season_type, timeout):
        """
        Respons√°vel por estruturar um dicion√°rio ordenado com as informa√ß√µes
        b√°sicas de uma requisi√ß√£o, este m√©todo pode ser utilizado para
        garantir o correto armazenamento de requisi√ß√µes com falhas, dado que
        recebe os par√¢metros b√°sicos da mesma como entrada e retorna
        um dicion√°rio contendo as chaves corretas a serem armazenadas
        no DataFrame definido pelo atributo self.error_data
        
        Par√¢metros
        ----------
        :param player_id:
            Id do jogador alvo da an√°lise.
            [type: int]
            
        :param player_name:
            Nome do jogador alvo da an√°lise.
            [type: string]
            
        :param season:
            Temporada definida na requisi√ß√£o executada.
            [type: string]
            
        :param season_type:
            Tipo de temporada na requisi√ß√£o executada. Normalmente definida 
            por "Regular Season" ou "Playoffs".
            [type: string]
            
        :param timeout:
            Tempo limite definido para a requisi√ß√£o executada.
            [type: int]
            
        Retorno
        -------
        :return error_dict:
            Dicion√°rio estruturado com os valores dos par√¢metros e chaves
            equivalentes as colunas da base armazenada de erros
            [type: dict]
        """
        
        return {'player_id': player_id, 
                'player_name': player_name, 
                'season': season, 
                'season_type': season_type,
                'timeout': timeout}
    
    # Armazenando requisi√ß√µes com status de falha
    def store_error_data(self, error_dict):
        """
        Utilizar um dicion√°rio estruturado com dados de requisi√ß√µes
        com falhas para armazenar as informa√ß√µes no atributo
        self.error_data da classe. Normalmente, os m√©todos da classe
        estruturam requisi√ß√µes com falhas a partir da execu√ß√£o
        pr√©via do m√©todo self.structure_error_data para garantir
        a correta organiza√ß√£o dos dados antes do empilhamento.
        
        Par√¢metros
        ----------
        :param error_dict:
            Dicion√°rio estruturado contendo os dados de requisi√ß√µes
            falhas a serem armazenadas no atributo self.error_data
            da classe. Para uma estrutura√ß√£o mais simples visando
            garantir a completude do processo, o m√©todo
            self.structure_error_data() pode ser executado previamente
            ao armazenamento.
            [type: dict]
        """
        
        # Verificando estrutura do dicion√°rio
        if self.error_cols != list(error_dict.keys()):
            try:
                error_dict = {k: v for k, v in zip(self.error_cols, error_dict.values())}
            except Exception as e:
                logger.error(f'Dicion√°rio de erro enviado de forma incorreta. Certifique-se de que a estrutura armazenada contenha as chaves definidas por "{self.error_cols}"')
                return
        
        # Construindo DataFrame com dados de requisi√ß√£o com falha
        try:
            request_error = pd.DataFrame(error_dict, index=[0])
            request_error.columns = self.error_cols

            # Armazenando erro e incrementando atributo da classe
            self.error_data = self.error_data.append(request_error)
        except Exception as e:
            logger.error(f'Erro ao armazenar requisi√ß√£o com falha. Exception: {e}')
            return        
        
    # Gerenciando reprocessamento de requisi√ß√µes com falhas
    def reprocess_error_requests(self, gamelog_data, error_strategy='infinity', 
                                 num_attempts=3, timeout_increase=10):
        """
        Executa e gerencia o reprocessamento de requisi√ß√µes com falhas
        a partir de dados armazenados no atributo self.error_data da
        classe. Como configura√ß√µes adicionais deste m√©todo, √© poss√≠vel
        definir a forma como as requisi√ß√µes ser√£o reprocessadas, seja
        atrav√©s de um n√∫mero finito de tentativas ou ent√£o em uma
        abordagem onde tentativas s√£o feitas at√© que o sucesso
        seja obtido.
        
        Par√¢metros
        ----------
        :param gamelog_data: 
            Dados de partidas extra√≠dos a partir de requisi√ß√µes
            bem sucessidas √† API da NBA.
            [type: pd.DataFrame]
            
        :param error_strategy:
            Define o tipo de estrat√©gia a ser utilizada no reprocessamento
            das requisi√ß√µes com falhas.
            
            *strategy='infinity': as tentativas de reprocessamento s√£o
            realizadas at√© que o sucesso completo seja obtido. Uma
            observa√ß√£o plaus√≠vel a ser feita √© que, nesse modo, a 
            execu√ß√£o do c√≥digo pode demorar de forma incalcul√°vel.
            
            *strategy='attempts': √© definido um limite de tentativas
            de reprocessamento dos dados. Nessa abordagem, evita-se que
            o codigo demande tempos elevados de execu√ß√£o, por√©m n√£o
            h√° a garantia que todos os dados ser√£o retornados para
            todos os jogadores.
            
            [type: string, default='infinity']
            
        :param num_attempts:
            N√∫mero de tentativas de reprocessamento em caso de 
            error_strategy='attempts'. Ao final de cada tentativa,
            a base de erros √© atualizada e verificada. Em caso de
            sucesso, a execu√ß√£o √© interrompida. Caso as tentativas
            se esgotem e ainda existam errors, os dados faltantes
            s√£o desconsiderados do retorno final.
            [type: int, default=3]
            
        :param timeout_increase:
            N√∫mero inteiro utilizado para incrementar o timeout da
            requisi√ß√£o a cada tentativa, propondo assim uma maior
            toler√¢ncia e uma maior chance de retorno dos dados a 
            cada nova requisi√ß√£o.
            [type: int, default=10]
            
        Retorno
        -------
        :return gamelog_data:
            Base de dados com requisi√ß√µes atualizadas ap√≥s as 
            tentativas de reprocessamento realizadas.
            [type: pd.DataFrame]
        """
        
        # Coletando requisi√ß√µes com falhas e resetando atributo da classe
        fail_requests = self.error_data.copy()
        self.error_data = pd.DataFrame(columns=self.error_cols)
        
        # Estrat√©gia de reprocessamento: n√∫mero finito de tentativas 
        if error_strategy == 'attempts':
            
            # Iterando sobre n√∫mero finito de tentativas
            for i in range(num_attempts):
                
                # Comunicando usu√°rio
                if self.verbose >= 1:
                    logger.debug(f'Tentativa n√∫mero {i + 1} de reprocessamento')
                    
                # Iterando sobre requisi√ß√µes com falhas
                for index, data in fail_requests.iterrows():
                    # Incrementando base de dados
                    gamelog_data = gamelog_data.append(self.player_gamelog_season(player_id=data['player_id'],
                                                                                  season=data['season'],
                                                                                  season_type_all_star=data['season_type'],
                                                                                  timeout=data['timeout'] + timeout_increase))

                # Copiando nova base de erros e verificando sucesso
                if len(self.error_data) == 0:
                    return gamelog_data
                else:
                    fail_requests = self.error_data.copy()
                    self.error_data = pd.DataFrame(columns=self.error_cols)

            return gamelog_data
        
        # Estrat√©gia de reprocessamento: la√ßo infinito
        elif error_strategy == 'infinity':
            
            # La√ßo infinito de itera√ß√£o at√© que o n√∫mero de erros seja 0
            i = 0
            while True:
                
                # Comunicando usu√°rio
                if self.verbose >= 1:
                    logger.debug(f'Tentativa n√∫mero {i + 1} de reprocessamento')
                    
                # Iterando sobre requisi√ß√µes com falhas
                for index, data in fail_requests.iterrows():
                    # Incrementando base de dados
                    gamelog_data = gamelog_data.append(self.player_gamelog_season(player_id=data['player_id'],
                                                                                  season=data['season'],
                                                                                  season_type_all_star=data['season_type'],
                                                                                  timeout=data['timeout'] + timeout_increase))
                
                # Copiando nova base de erros e verificando sucesso
                if len(self.error_data) == 0:
                    return gamelog_data
                else:
                    fail_requests = self.error_data.copy()
                    self.error_data = pd.DataFrame(columns=self.error_cols)
                    i += 1

            return gamelog_data
        
    # Identificando jogadores utilizando o id
    def player_identification(self, player_id, player_id_key='full_name'):
        """
        Prop√µe uma identifica√ß√£o detalhada de um jogador a partir de seu id.
        Para isso, utiliza basicamente o m√©todo find_player_by_id() do
        atributo players do m√≥dulo est√°tico nba_api.stats.static.
        
        Par√¢metros
        ----------
        :param player_id:
            Id do jogador a ser identificado em detalhes.
            [type: int]
            
        :param player_id_key:
            Chave de identifica√ß√£o utilizada no dicion√°rio de retorno
            do m√©todo find_players_by_id.
            [type: string, default='full_name']
            
        :return player_id:
            Nova identifica√ß√£o do usu√°rio.
            [type: string]
        """
        
        try:
            players_dict = players.find_player_by_id(player_id=player_id)
            if player_id_key not in players_dict.keys():
                logger.warning(f'Chave "{player_id_key}" n√£o dispon√≠vel. Escolha entre as op√ß√µes: {list(players_dict.keys())}. Retornando id de entrada.')
                return player_id
            else:
                return players_dict[player_id_key]
        except Exception as e:
            logger.error(f'Erro ao encontrar jogador {player_id}. Retornando id de entrada')
            return player_id
    
    # Gerando lista formatada de temporadas a serem consideradas nas extra√ß√µes hist√≥ricas
    def get_seasons_list(self, from_year, to_year):
        """
        Retorna uma lista formatada com um intervalo de temporadas
        a serem utilizadas em requisi√ß√µes hist√≥ricas para um determinado
        jogador.
        
        Par√¢metros
        ----------
        :param from_year: 
            Ano de in√≠cio da lista hist√≥rica de temporadas.
            [type: int]
            
        :param to_year:
            Ano final da lista hist√≥rica de temporadas.
            [type: int]
            
        Retorno
        -------
        :return seasons_list:
            Lista configurada com cada temporada a ser utilizada 
            nas requisi√ß√µes no formato 'yyyy-yy'.
            Exemplo: ['2018-19', '2019-20', '2020-21']
            [type: list]
        """
        return [str(year - 1) + '-' + str(year)[-2:] for year in range(from_year, to_year + 1)]        
    
    # Retornando gamelog de um √∫nico jogador em uma √∫nica temporada (tipo espec√≠fico)
    def player_gamelog_season(self, player_id, season, season_type_all_star='Regular Season', 
                              timeout=60, **kwargs):
        """
        Realiza uma requisi√ß√£o unit√°ria de um √∫nico jogador, 
        em uma √∫nica temporada e considerando um √∫nico tipo de 
        temporada (regular ou playoffs). Eventualmente, este
        m√©todo poder√° ser utilizado por outros m√©todos da
        classe para requisi√ß√µes dentro de la√ßos de itera√ß√£o.
        
        Em caso de falhas na requisi√ß√£o, as informa√ß√µes b√°sicas
        s√£o armazenadas no atributo self.error_data da classe.
        
        Par√¢metros
        ----------
        :param player_id:
            Id do jogador alvo da requisi√ß√£o.
            [type: int]
            
        :param season:
            Temporada alvo da requisi√ß√£o.
            [type: int]
            
        :param season_type:
            Tipo de temporada alvo da extra√ß√£o.
            [type: string, default='Regular Season']
            
        :param timeout:
            Limite de tempo, em segundos, a ser considerado para
            cada requisi√ß√£o.
            [type: int, default=60]
            
        **kwargs
        --------
        :arg player_log_id:
            Identificador customizado do jogador para fins
            de armazenamento de logs e erros.
            [type: string, default=self.player_identification(player_id, 'full_name')]
            
        Retorno
        -------
        :return df_gamelog:
            Retorno da requisi√ß√£o unit√°ria para o jogador alvo
            em um formato DataFrame.
            [type: pd.DataFrame]
        """
        
        try:
            # Retornando gamelog de um jogador espec√≠fico em um tipo de temporada espec√≠fica
            gamelog_season = playergamelog.PlayerGameLog(player_id=player_id,
                                                         season=season,
                                                         season_type_all_star=season_type_all_star,
                                                         timeout=timeout)

            # Coletando dados da requisi√ß√£o em formato DataFrame
            df_gamelog = gamelog_season.player_game_log.get_data_frame()
            df_gamelog['SEASON'] = season
            df_gamelog['SEASON_TYPE'] = season_type_all_star

            # Transformando coluna de data na base
            df_gamelog['GAME_DATE'] = pd.to_datetime(df_gamelog['GAME_DATE'])

            return df_gamelog

        except Exception as e:
            # Extraindo informa√ß√µes detalhadas e armazenando no atributo da classe
            player_log_id = kwargs['player_log_id'] if 'player_log_id' in kwargs else self.player_identification(player_id, 'full_name')
            error_dict = self.structure_error_data(player_id, player_log_id, season, season_type_all_star, timeout)
            self.store_error_data(error_dict)
            
            # Comunicando log de acordo com o atributo verbose
            if self.verbose > 1:
                logger.error(f'Erro ao retornar dados do jogador {player_log_id} em {season_type_all_star} {season}. Exception: {e}')
            
            return None
    
    # Retornando gamelog de um √∫nico jogador em uma √∫nica temporada (Regular + Playoffs)
    def player_gamelog_season_complete(self, player_id, season, season_types=['Regular Season', 'Playoffs'],
                                       timeout=60):
        """
        Realiza m√∫ltiplas requi√ß√µes para extrair dados de um
        jogador considerando diferentes tipos de temporadas.
        Com esse m√©todo √© poss√≠vel, por exemplo, obter dados
        de partidas de um jogador, em uma temporada √∫nica,
        considerando dados de playoffs e temporada regular.
        
        Par√¢metros
        ----------
        :param player_id:
            Id do jogador alvo da requisi√ß√£o.
            [type: int]
            
        :param season:
            Temporada alvo da requisi√ß√£o.
            [type: int]
            
        :param season_types:
            Tipos de temporada alvo da extra√ß√£o.
            [type: list, default=['Regular Season', 'Playoffs']]
            
        :param timeout:
            Limite de tempo, em segundos, a ser considerado para
            cada requisi√ß√£o.
            [type: int, default=60]
            
        Retorno
        -------
        :return df_season_gamelog:
            Retorno da requisi√ß√£o para o jogador alvo
            em um formato DataFrame.
            [type: pd.DataFrame]"""
        
        # Ajustando par√¢metro de season em caso default - considera temporada atual
        if season == 'default':
            season = str(datetime.now().year - 1) + '-' + str(datetime.now().year)[-2:]
        
        # Inicializando vari√°veis de controle
        df_season_gamelog = pd.DataFrame()

        # Iterando sobre a lista de tipos de temporadas
        for season_type in season_types:
            df_season_gamelog = df_season_gamelog.append(self.player_gamelog_season(player_id=player_id, 
                                                                                    season=season,
                                                                                    season_type_all_star=season_type,
                                                                                    timeout=timeout))

        # Ordenando dados por data da partida e retornando DataFrame com √≠ndice ajustado
        df_season_gamelog.sort_values(by='GAME_DATE', ascending=False, inplace=True)
        
        return df_season_gamelog.reset_index(drop=True)
    
    # Retornando gamelog de um √∫nico jogador em todas as temporadas (tipo espec√≠fico)
    def player_gamelog_all_seasons(self, player_id, player_log_key='full_name',
                                   season_type_all_star='Regular Season', timeout=60):
        
        """
        Realiza m√∫ltiplas requi√ß√µes para extrair dados de um
        jogador considerando todas as temporadas dispon√≠veis
        para o mesmo em um √∫nico tipo de temporada.
        Com esse m√©todo √© poss√≠vel, por exemplo, obter dados
        de partidas de um jogador, em todo seu per√≠odo de
        atividade na NBA, considerando um √∫nico tipo de temporada
        (regular ou playoffs).
        
        Par√¢metros
        ----------
        :param player_id:
            Id do jogador alvo da requisi√ß√£o.
            [type: int]
            
        :param player_log_key:
            Chave de identifica√ß√£o utilizada no dicion√°rio de retorno
            do m√©todo find_players_by_id.
            [type: string, default='full_name']
            
        :param season_type_all_star:
            Tipos de temporada alvo da extra√ß√£o.
            [type: string, default='Regular Season']
            
        :param timeout:
            Limite de tempo, em segundos, a ser considerado para
            cada requisi√ß√£o.
            [type: int, default=60]
            
        Retorno
        -------
        :return df_gamelog:
            Retorno da requisi√ß√£o para o jogador alvo
            em um formato DataFrame.
            [type: pd.DataFrame]"""
        
        # Criando DataFrame vazio para armazenar requisi√ß√µes
        df_gamelog = pd.DataFrame()
        player_log_id = self.player_identification(player_id=player_id, player_id_key=player_log_key)
        
        # Extraindo lista de temporadas e comunicando usu√°rio
        attempts = 1
        while True:
            try:
                # Requisi√ß√£o para intervalo de temporadas v√°lidas do jogador
                player_info = commonplayerinfo.CommonPlayerInfo(player_id, timeout=timeout).common_player_info.get_data_frame()
                from_year = player_info['FROM_YEAR'].values[0]
                to_year = player_info['TO_YEAR'].values[0]
                break
            except Exception as e:
                if attempts % 5 == 0:
                    logger.warning(f'{attempts} realizadas para extrair temporadas v√°lidas de {player_log_id}')
                attempts += 1
                continue
            
        # Iterando sobre todas as temporadas do jogador
        logger.debug(f'Extraindo hist√≥rico de partidas de {player_log_id} em {season_type_all_star} entre {from_year} e {to_year}')
        for season in self.get_seasons_list(from_year, to_year):
            df_gamelog = df_gamelog.append(self.player_gamelog_season(player_id=player_id,
                                                                      season=season,
                                                                      season_type_all_star=season_type_all_star,
                                                                      timeout=timeout))
        return df_gamelog
          
    # Retornando gamelog de um √∫nico jogador em todas as temporadas (Regular + Playoffs)
    def player_gamelog_all_seasons_complete(self, player_id, season_types=['Regular Season', 'Playoffs'], 
                                            player_log_key='full_name', timeout=60, reprocess=True, **kwargs):
        
        """
        Realiza m√∫ltiplas requi√ß√µes para extrair dados de um
        jogador considerando todas as temporadas dispon√≠veis
        para o mesmo em todos os tipos de temporadas.
        Com esse m√©todo √© poss√≠vel, por exemplo, obter dados
        de partidas de um jogador, em todo seu per√≠odo de
        atividade na NBA, considerando temporada regular
        e playoffs.
        
        Par√¢metros
        ----------
        :param player_id:
            Id do jogador alvo da requisi√ß√£o.
            [type: int]
            
        :param player_log_key:
            Chave de identifica√ß√£o utilizada no dicion√°rio de retorno
            do m√©todo find_players_by_id.
            [type: string, default='full_name']
            
        :param season_type_all_star:
            Tipos de temporada alvo da extra√ß√£o.
            [type: string, default=['Regular Season', 'Playoffs']]
            
        :param timeout:
            Limite de tempo, em segundos, a ser considerado para
            cada requisi√ß√£o.
            [type: int, default=60]
            
        :param reprocess:
            Flag para guiar o reprocessamento dos dados
            gerados por requisi√ß√µes falhas.
            [type: bool, default=True]
            
        **kwargs
        --------
        :arg error_strategy:
            Define o tipo de estrat√©gia a ser utilizada no reprocessamento
            das requisi√ß√µes com falhas.
            
            *strategy='infinity': as tentativas de reprocessamento s√£o
            realizadas at√© que o sucesso completo seja obtido. Uma
            observa√ß√£o plaus√≠vel a ser feita √© que, nesse modo, a 
            execu√ß√£o do c√≥digo pode demorar de forma incalcul√°vel.
            
            *strategy='attempts': √© definido um limite de tentativas
            de reprocessamento dos dados. Nessa abordagem, evita-se que
            o codigo demande tempos elevados de execu√ß√£o, por√©m n√£o
            h√° a garantia que todos os dados ser√£o retornados para
            todos os jogadores.
            
            [type: string, default='infinity']
            
        :arg num_attempts:
            N√∫mero de tentativas de reprocessamento em caso de 
            error_strategy='attempts'. Ao final de cada tentativa,
            a base de erros √© atualizada e verificada. Em caso de
            sucesso, a execu√ß√£o √© interrompida. Caso as tentativas
            se esgotem e ainda existam errors, os dados faltantes
            s√£o desconsiderados do retorno final.
            [type: int, default=3]
            
        :arg timeout_increase:
            N√∫mero inteiro utilizado para incrementar o timeout da
            requisi√ß√£o a cada tentativa, propondo assim uma maior
            toler√¢ncia e uma maior chance de retorno dos dados a 
            cada nova requisi√ß√£o.
            [type: int, default=10]
            
        Retorno
        -------
        :return df_all_seasons:
            Retorno da requisi√ß√£o para o jogador alvo
            em um formato DataFrame.
            [type: pd.DataFrame]"""
        
        # Inicializando vari√°veis de controle
        df_all_seasons = pd.DataFrame()
        player_log_id = self.player_identification(player_id=player_id, player_id_key=player_log_key)
        
        # Extraindo lista de temporadas e comunicando usu√°rio
        attempts = 0
        while True:
            try:
                # Requisi√ß√£o para intervalo de temporadas v√°lidas do jogador
                player_info = commonplayerinfo.CommonPlayerInfo(player_id, timeout=timeout).common_player_info.get_data_frame()
                from_year = player_info['FROM_YEAR'].values[0]
                to_year = player_info['TO_YEAR'].values[0]
                
                # Extraindo lista de temporadas dispon√≠veis para o jogador
                player_seasons = self.get_seasons_list(from_year, to_year)
                break
            except Exception as e:
                attempts += 1
                if attempts % 5 == 0:
                    logger.warning(f'{attempts + 1} realizadas para extrair temporadas v√°lidas de {player_log_id}')
                continue
            
        # Iterando sobre todas as temporadas Regualres e Playoffs
        logger.debug(f'Extraindo hist√≥rico de partidas de {player_log_id} entre {from_year} e {to_year}')
        for season in player_seasons:
            for season_type in season_types:
                df_all_seasons = df_all_seasons.append(self.player_gamelog_season(player_id=player_id,
                                                                                  player_log_id=player_log_id,
                                                                                  season=season,
                                                                                  season_type_all_star=season_type,
                                                                                  timeout=timeout))        
        
        # Extraindo argumentos adicionais e verificando necessidade de reprocessamento
        total_requests = len(player_seasons) * len(season_types)
        
        if reprocess and len(self.error_data) > 0:
            pct_success = str(round(100 * (total_requests - len(self.error_data)) / total_requests, 1)) + '%'
            logger.warning(f'Dados de {player_log_id} retornados com {pct_success} de sucesso. Iniciando reprocessamento')
            
            # Extra√ß√£o de par√¢metros de reprocessamento
            error_strategy = kwargs['error_strategy'] if 'error_strategy' in kwargs else 'infinity'
            num_attempts = kwargs['num_attempts'] if 'num_attempts' in kwargs else 3
            timeout_increase = kwargs['timeout_increase'] if 'timeout_increase' in kwargs else 10
            
            # Atualizando DataFrame com novos dados ap√≥s reprocessamento
            df_all_seasons = self.reprocess_error_requests(gamelog_data=df_all_seasons, 
                                                           error_strategy=error_strategy,
                                                           num_attempts=num_attempts,
                                                           timeout_increase=timeout_increase)
                
        # Ordenando dados por data da partida e retornando DataFrame com √≠ndice ajustado
        final_pct_success = str(round(100 * (total_requests - len(self.error_data)) / total_requests, 1)) + '%'
        logger.info(f'Processo finalizado para {player_log_id} com {final_pct_success} de sucesso')
        df_all_seasons.sort_values(by='GAME_DATE', ascending=False, inplace=True)
               
        return df_all_seasons.reset_index(drop=True)
    
    # Retornando gamelog de todos os jogadores ativos em todas as temporadas (Regular + Playoffs)
    def active_players_gamelog(self, season_types=['Regular Season', 'Playoffs'], 
                               player_log_key='full_name', timeout=60, reprocess=True, **kwargs):
        """
        Realiza m√∫ltiplas requisi√ß√µes para todos os jogadores ativos
        da liga para extrair hist√≥rico de partidas em todas as 
        respectivas temporadas v√°lidas para cada jogador, considerando
        temporada regular e playoffs.
        
        Par√¢metros
        ----------            
        :param player_log_key:
            Chave de identifica√ß√£o utilizada no dicion√°rio de retorno
            do m√©todo find_players_by_id.
            [type: string, default='full_name']
            
        :param season_type_all_star:
            Tipos de temporada alvo da extra√ß√£o.
            [type: string, default=['Regular Season', 'Playoffs']]
            
        :param timeout:
            Limite de tempo, em segundos, a ser considerado para
            cada requisi√ß√£o.
            [type: int, default=60]
            
        :param reprocess:
            Flag para guiar o reprocessamento dos dados
            gerados por requisi√ß√µes falhas.
            [type: bool, default=True]
            
        **kwargs
        --------
        :arg error_strategy:
            Define o tipo de estrat√©gia a ser utilizada no reprocessamento
            das requisi√ß√µes com falhas.
            
            *strategy='infinity': as tentativas de reprocessamento s√£o
            realizadas at√© que o sucesso completo seja obtido. Uma
            observa√ß√£o plaus√≠vel a ser feita √© que, nesse modo, a 
            execu√ß√£o do c√≥digo pode demorar de forma incalcul√°vel.
            
            *strategy='attempts': √© definido um limite de tentativas
            de reprocessamento dos dados. Nessa abordagem, evita-se que
            o codigo demande tempos elevados de execu√ß√£o, por√©m n√£o
            h√° a garantia que todos os dados ser√£o retornados para
            todos os jogadores.
            
            [type: string, default='infinity']
            
        :arg num_attempts:
            N√∫mero de tentativas de reprocessamento em caso de 
            error_strategy='attempts'. Ao final de cada tentativa,
            a base de erros √© atualizada e verificada. Em caso de
            sucesso, a execu√ß√£o √© interrompida. Caso as tentativas
            se esgotem e ainda existam errors, os dados faltantes
            s√£o desconsiderados do retorno final.
            [type: int, default=3]
            
        :arg timeout_increase:
            N√∫mero inteiro utilizado para incrementar o timeout da
            requisi√ß√£o a cada tentativa, propondo assim uma maior
            toler√¢ncia e uma maior chance de retorno dos dados a 
            cada nova requisi√ß√£o.
            [type: int, default=10]
            
        Retorno
        -------
        :return df_all_seasons:
            Retorno da requisi√ß√£o para o jogador alvo
            em um formato DataFrame.
            [type: pd.DataFrame]"""
        
        # Inicializando vari√°veis de controle
        complete_gamelog = pd.DataFrame()
        error_strategy = kwargs['error_strategy'] if 'error_strategy' in kwargs else 'infinity'
        num_attempts = kwargs['num_attempts'] if 'num_attempts' in kwargs else 3
        timeout_increase = kwargs['timeout_increase'] if 'timeout_increase' in kwargs else 10
        
        # Iterando sobre jogadores ativos
        for player_id in nba_gamelog.active_players['id'].values:
            player_log_id = self.player_identification(player_id=player_id, player_id_key=player_log_key)
            complete_gamelog = complete_gamelog.append(self.player_gamelog_all_seasons_complete(player_id=player_id,
                                                                                                season_types=season_types,
                                                                                                player_log_key=player_log_key,
                                                                                                timeout=timeout,
                                                                                                reprocess=reprocess,
                                                                                                error_strategy=error_strategy,
                                                                                                num_attempts=num_attempts,
                                                                                                timeout_increase=timeout_increase))
        return complete_gamelog
    
```
